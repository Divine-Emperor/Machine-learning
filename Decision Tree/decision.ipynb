```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# ðŸ“Œ Step 1: Create a simple dataset
data = {
    'weather': ['sunny', 'rainy', 'overcast', 'sunny', 'rainy', 'overcast', 'sunny', 'rainy'],
    'temperature': ['hot', 'cold', 'mild', 'hot', 'cold', 'mild', 'hot', 'cold'],
    'go_outside': ['yes', 'no', 'yes', 'yes', 'no', 'yes', 'yes', 'no']  # Target variable
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# ðŸ“Œ Step 2: Convert categorical features into numbers
df['weather'] = df['weather'].replace({'sunny': 0, 'rainy': 1, 'overcast': 2})
df['temperature'] = df['temperature'].replace({'hot': 0, 'cold': 1, 'mild': 2})
df['go_outside'] = df['go_outside'].replace({'yes': 1, 'no': 0})
```

```python
df
```

```python
# ðŸ“Œ Step 3: Features & Target
X = df[['weather', 'temperature']]
y = df['go_outside']
```

```python
# ðŸ“Œ Step 4: Train Decision Tree
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X, y)

# ðŸ“Œ Step 5: Visualize Decision Tree
plt.figure(figsize=(10, 4))
plot_tree(clf, feature_names=X.columns, class_names=['No', 'Yes'], filled=True)
plt.show()
```
